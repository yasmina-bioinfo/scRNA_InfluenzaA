```python
path = "data_raw/geo_mtx"
print(os.path.exists(path))
print(os.listdir(path)[:5])
```

```python
import scanpy as sc
import pandas as pd

# Read matrix (genes x cells in Cell Ranger)
X = sc.read_mtx(f"{path}/matrix.mtx").X

# Read barcodes (cells)
barcodes = pd.read_csv(f"{path}/barcodes.tsv", header=None, sep="\t")[0].astype(str).tolist()

# Read genes (old Cell Ranger format: genes.tsv = gene_id, gene_name)
genes = pd.read_csv(f"{path}/genes.tsv", header=None, sep="\t")
gene_ids = genes[0].astype(str).tolist()
gene_names = genes[1].astype(str).tolist()

# Build AnnData (cells x genes)
adata = sc.AnnData(X.T)
adata.obs_names = barcodes
adata.var["gene_ids"] = gene_ids
adata.var_names = gene_names

print(adata)
```

```python
adata.write("results/adata_raw.h5ad")
```
